Create dataframe with missing values

```python
raw_data = {'First_name':['Ricky',np.nan,'John','Rim','Ravi','Tina'],
            'Last_name':['singh',np.nan,'kang','mital','rath','dev'],
            'age':[22,np.nan,23,24,25,26],
            'Sex':['M',np.nan,'M','M','M','F'],
             'unit_1':[1,np.nan,np.nan,3,4,4],
             'unit_2':[3,np.nan,np.nan,4,3,3]}
df = pd.DataFrame(raw_data,columns = ['First_name','Last_name','age','Sex','unit_1','unit_2'])
df
```

Drop missing observations

```python
df.dropna()
```


Drop rows where all cells in the row is NA

```python
df
```

```python
#Row that has all NaN will be removed
df.dropna(how='all')
```


create a new column full of missing values

```python
df['Final_score'] = np.nan
df
```

```python
# drop column if they only contain missing values
df.dropna(axis = 1,how = 'all')
```

```python
df.fillna(0)
```

```python
df
```


FILL IN MISSING IN UNIT_1 WITH THE MEAN VALUE OF UNIT_1

```python
df['unit_1'].fillna(df['unit_1'].mean(),inplace = True)
df
```


FILL IN MISSING IN POST TEST SCORE WITH EACH SEX MEAN VALUE OF POST TEST SCORE

```python
df['unit_2'].fillna(df.groupby('Sex')['unit_2'].transform('mean'),inplace = True)
df
```


SELECT SOME RAWS BUT IGNORE THE MISSING DATA POINT

```python
df[df['age'].notnull() & df['Sex'].notnull()]
```

```python
df
```

```python
#Back fill
df.fillna(method = 'bfill')
```

```python
df
```

```python
df.fillna(method = 'ffill',inplace = True)
```

```python
df
```

```python
df['Final_score'].fillna(df['unit_1'] + df['unit_2'],inplace = True)
```


TITANIC DATA SET

```python
titanic_train = pd.read_csv('https://raw.githubusercontent.com/training-ml/Files/main/titanic_train.csv')
titanic_train.head()
```


HOW TO CHECK UNIQUE VALUES OF ANY COLUMN

```python
titanic_train['Survived'].unique() #similarly you can check any column unique value
```

 HOW TO RENAME CATEGORIES USING PANDAS METHOD

```python
new_survived = pd.Categorical(titanic_train['Survived'])
new_survived = new_survived.rename_categories(['Dead','Survived'])
new_survived.describe()
```

```python
titanic_train['Pclass'].unique()
```

```python
new_Pclass = pd.Categorical(titanic_train['Pclass'],ordered = True)
new_Pclass = new_Pclass.rename_categories(['class1','class2','class3'])
new_Pclass.describe()
```

```python
#Reassign new name to Pclass
titanic_train['Pclass'] = new_Pclass
```

```python
titanic_train.head()
```

```python
titanic_train['Cabin'].unique() #check unique cabin
```

```python
char_cabin = titanic_train['Cabin'].astype(str) # convert data to str
# List comprehension(creating list using anoyher type of string,list etc)
new_cabin = np.array([cabin[0] for cabin in char_cabin])# Take first letter
new_cabin = pd.Categorical(new_cabin)
new_cabin.describe()
```

```python
titanic_train['Cabin'] = new_cabin
```

```python
titanic_train.head(10)
```




FIND ALL THE PASSENGERS WHOSE AGE IS MISSING

```python
## np.where gives the index of those missing data
missing = np.where(titanic_train['Age'].isnull() == True) # Find all nulls in column age
missing
```

```python
# Using the index locate passangers
titanic_train.loc[missing]
```



FIND THE MAXIMUM FARE




```python
INDEX = np.where(titanic_train['Fare'] == max(titanic_train['Fare']))
INDEX
```

```python
titanic_train.loc[INDEX]
```



FIND THE FAMILY WHO HAS HIGHEST MEMBERS

```python
# Sibsp and parch are siblings and parents/guardian respectively(add both of them and create column)
titanic_train['Family'] = titanic_train['SibSp'] + titanic_train['Parch']
most_family = np.where(titanic_train['Family'] == max(titanic_train['Family']))
titanic_train.loc[most_family]
```
